```python
# afstandsmatrix op basis van GPS-coördinaten

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

n = len(locations)
distance_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        distance_matrix[i, j] = haversine(
            locations.loc[i, 'Lat'],
            locations.loc[i, 'Long'],
            locations.loc[j, 'Lat'],
            locations.loc[j, 'Long']
        )

# OR-Tools werkt liever met integers
distance_matrix = (distance_matrix * 1000).astype(int)  # meters
```

## MODEL 1

```python
import time
import pandas as pd
from ortools.constraint_solver import pywrapcp, routing_enums_pb2
```

```python
TIME_LIMIT_SECONDS = 30

first_solution_strategies = {
    "PATH_CHEAPEST_ARC": routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC,
    "PATH_MOST_CONSTRAINED_ARC": routing_enums_pb2.FirstSolutionStrategy.PATH_MOST_CONSTRAINED_ARC,
    "EVALUATOR_STRATEGY": routing_enums_pb2.FirstSolutionStrategy.EVALUATOR_STRATEGY,
    "SAVINGS": routing_enums_pb2.FirstSolutionStrategy.SAVINGS,
    "SWEEP": routing_enums_pb2.FirstSolutionStrategy.SWEEP,
    "CHRISTOFIDES": routing_enums_pb2.FirstSolutionStrategy.CHRISTOFIDES,
    "ALL_UNPERFORMED": routing_enums_pb2.FirstSolutionStrategy.ALL_UNPERFORMED,
    "BEST_INSERTION": routing_enums_pb2.FirstSolutionStrategy.BEST_INSERTION,
    "PARALLEL_CHEAPEST_INSERTION": routing_enums_pb2.FirstSolutionStrategy.PARALLEL_CHEAPEST_INSERTION,
    "LOCAL_CHEAPEST_INSERTION": routing_enums_pb2.FirstSolutionStrategy.LOCAL_CHEAPEST_INSERTION,
    "GLOBAL_CHEAPEST_ARC": routing_enums_pb2.FirstSolutionStrategy.GLOBAL_CHEAPEST_ARC,
    "LOCAL_CHEAPEST_ARC": routing_enums_pb2.FirstSolutionStrategy.LOCAL_CHEAPEST_ARC,
    "FIRST_UNBOUND_MIN_VALUE": routing_enums_pb2.FirstSolutionStrategy.FIRST_UNBOUND_MIN_VALUE,
}

local_search_metaheuristics = {
    "GREEDY_DESCENT": routing_enums_pb2.LocalSearchMetaheuristic.GREEDY_DESCENT,
    "GUIDED_LOCAL_SEARCH": routing_enums_pb2.LocalSearchMetaheuristic.GUIDED_LOCAL_SEARCH,
    "SIMULATED_ANNEALING": routing_enums_pb2.LocalSearchMetaheuristic.SIMULATED_ANNEALING,
    "TABU_SEARCH": routing_enums_pb2.LocalSearchMetaheuristic.TABU_SEARCH,
    "GENERIC_TABU_SEARCH": routing_enums_pb2.LocalSearchMetaheuristic.GENERIC_TABU_SEARCH,
}
```

```python
# FUNCTIE: MODEL OPBOUWEN
def build_routing_model(distance_matrix, num_vehicles, depot=0):
    """
    Bouwt steeds een vers RoutingIndexManager + RoutingModel.
    """
    manager = pywrapcp.RoutingIndexManager(
        len(distance_matrix),
        num_vehicles,
        [depot] * num_vehicles,   # start depot per voertuig
        [depot] * num_vehicles    # eind depot per voertuig
    )

    routing = pywrapcp.RoutingModel(manager)

    def distance_callback(from_index, to_index):
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return int(distance_matrix[from_node][to_node])

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    return manager, routing
```

```python
# FUNCTIE: TOTALE AFSTAND BEREKENEN
def calculate_total_distance(solution, routing, num_vehicles):
    total_distance = 0

    for vehicle_id in range(num_vehicles):
        index = routing.Start(vehicle_id)
        while not routing.IsEnd(index):
            next_index = solution.Value(routing.NextVar(index))
            total_distance += routing.GetArcCostForVehicle(index, next_index, vehicle_id)
            index = next_index

    return total_distance
```

```python
# FUNCTIE: RESULTATEN UIT ROUTES HALEN
def extract_routes(solution, manager, routing, num_vehicles, locations):
    totaal_afstand = 0
    routes_output = []

    for vehicle_id in range(num_vehicles):
        index = routing.Start(vehicle_id)
        route_nodes = []
        route_adressen = []
        route_distance = 0

        while not routing.IsEnd(index):
            node = manager.IndexToNode(index)
            route_nodes.append(node)
            route_adressen.append(locations.loc[node, 'VisitingAddress'])

            previous_index = index
            index = solution.Value(routing.NextVar(index))
            route_distance += routing.GetArcCostForVehicle(previous_index, index, vehicle_id)

        # einddepot toevoegen
        end_node = manager.IndexToNode(index)
        route_nodes.append(end_node)
        route_adressen.append(locations.loc[end_node, 'VisitingAddress'])

        totaal_afstand += route_distance

        routes_output.append({
            'vehicle_id': vehicle_id,
            'route_nodes': route_nodes,
            'route_adressen': route_adressen,
            'afstand_meters': route_distance,
            'afstand_km': round(route_distance / 1000, 2)
        })

    return totaal_afstand, routes_output
```

```python
# FUNCTIE: ALLE COMBINATIES TESTEN
def compare_strategies(distance_matrix, num_vehicles,
                       first_solution_strategies,
                       local_search_metaheuristics,
                       time_limit_seconds=30):
    results = []

    for fs_name, fs_value in first_solution_strategies.items():
        for ls_name, ls_value in local_search_metaheuristics.items():

            # Altijd een vers model bouwen
            manager, routing = build_routing_model(distance_matrix, num_vehicles, depot=0)

            search_parameters = pywrapcp.DefaultRoutingSearchParameters()
            search_parameters.first_solution_strategy = fs_value
            search_parameters.local_search_metaheuristic = ls_value
            search_parameters.time_limit.seconds = time_limit_seconds

            start = time.time()
            solution = routing.SolveWithParameters(search_parameters)
            runtime = time.time() - start

            if solution:
                total_distance = calculate_total_distance(solution, routing, num_vehicles)

                results.append({
                    "first_solution_strategy": fs_name,
                    "local_search_metaheuristic": ls_name,
                    "solution_found": True,
                    "total_distance": total_distance,
                    "runtime_sec": round(runtime, 2),
                })
            else:
                results.append({
                    "first_solution_strategy": fs_name,
                    "local_search_metaheuristic": ls_name,
                    "solution_found": False,
                    "total_distance": None,
                    "runtime_sec": round(runtime, 2),
                })

    df_results = pd.DataFrame(results)

    # Sorteer: eerst oplossingen, dan laagste afstand, dan snelste runtime
    df_results = df_results.sort_values(
        by=["solution_found", "total_distance", "runtime_sec"],
        ascending=[False, True, True]
    ).reset_index(drop=True)

    return df_results
```

```python
# FUNCTIE: BESTE PARAMETERS KIEZEN
def get_best_parameters(df_results, allow_zero_distance=False):
    """
    Kiest de beste combinatie uit de resultaten.
    Standaard worden verdachte oplossingen met total_distance <= 0 niet gekozen.
    """

    df_valid = df_results[df_results["solution_found"] == True].copy()

    # Verwijder lege waarden
    df_valid = df_valid[df_valid["total_distance"].notna()]

    # Filter verdachte 0-afstand oplossingen eruit
    if not allow_zero_distance:
        df_valid = df_valid[df_valid["total_distance"] > 0]

    if df_valid.empty:
        raise ValueError(
            "Geen geldige oplossing gevonden. "
            "Controleer je distance_matrix, aantal voertuigen en modelinstellingen."
        )

    best_row = df_valid.sort_values(
        by=["total_distance", "runtime_sec"],
        ascending=[True, True]
    ).iloc[0]

    return best_row["first_solution_strategy"], best_row["local_search_metaheuristic"]
```

```python
# BESTE COMBINATIE KIEZEN
# =========================================================
best_fs, best_ls = get_best_parameters(df_results, allow_zero_distance=False)

print("\nBeste combinatie voor Model 1:")
print(f"FirstSolutionStrategy: {best_fs}")
print(f"LocalSearchMetaheuristic: {best_ls}")
```

## MODEL 2

```python
TIME_LIMIT_SECONDS = 30
MAX_ADRESSEN_PER_BUS = 15

first_solution_strategies = {
    "PATH_CHEAPEST_ARC": routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC,
    "PATH_MOST_CONSTRAINED_ARC": routing_enums_pb2.FirstSolutionStrategy.PATH_MOST_CONSTRAINED_ARC,
    "EVALUATOR_STRATEGY": routing_enums_pb2.FirstSolutionStrategy.EVALUATOR_STRATEGY,
    "SAVINGS": routing_enums_pb2.FirstSolutionStrategy.SAVINGS,
    "SWEEP": routing_enums_pb2.FirstSolutionStrategy.SWEEP,
    "CHRISTOFIDES": routing_enums_pb2.FirstSolutionStrategy.CHRISTOFIDES,
    "ALL_UNPERFORMED": routing_enums_pb2.FirstSolutionStrategy.ALL_UNPERFORMED,
    "BEST_INSERTION": routing_enums_pb2.FirstSolutionStrategy.BEST_INSERTION,
    "PARALLEL_CHEAPEST_INSERTION": routing_enums_pb2.FirstSolutionStrategy.PARALLEL_CHEAPEST_INSERTION,
    "LOCAL_CHEAPEST_INSERTION": routing_enums_pb2.FirstSolutionStrategy.LOCAL_CHEAPEST_INSERTION,
    "GLOBAL_CHEAPEST_ARC": routing_enums_pb2.FirstSolutionStrategy.GLOBAL_CHEAPEST_ARC,
    "LOCAL_CHEAPEST_ARC": routing_enums_pb2.FirstSolutionStrategy.LOCAL_CHEAPEST_ARC,
    "FIRST_UNBOUND_MIN_VALUE": routing_enums_pb2.FirstSolutionStrategy.FIRST_UNBOUND_MIN_VALUE,
}

local_search_metaheuristics = {
    "GREEDY_DESCENT": routing_enums_pb2.LocalSearchMetaheuristic.GREEDY_DESCENT,
    "GUIDED_LOCAL_SEARCH": routing_enums_pb2.LocalSearchMetaheuristic.GUIDED_LOCAL_SEARCH,
    "SIMULATED_ANNEALING": routing_enums_pb2.LocalSearchMetaheuristic.SIMULATED_ANNEALING,
    "TABU_SEARCH": routing_enums_pb2.LocalSearchMetaheuristic.TABU_SEARCH,
    "GENERIC_TABU_SEARCH": routing_enums_pb2.LocalSearchMetaheuristic.GENERIC_TABU_SEARCH,
}
```

```python
# depot = 0, alle andere adressen krijgen vraag 1
demands = [0] + [1] * (len(locations) - 1)

# elk voertuig mag max 15 adressen doen
vehicle_capacities = [MAX_ADRESSEN_PER_BUS] * num_vehicles
```

```python
# MODEL 2
def build_capacity_model(distance_matrix, num_vehicles, demands, vehicle_capacities, depot=0):
    manager = pywrapcp.RoutingIndexManager(
        len(distance_matrix),
        num_vehicles,
        [depot] * num_vehicles,
        [depot] * num_vehicles
    )

    routing = pywrapcp.RoutingModel(manager)

    # afstand callback
    def distance_callback(from_index, to_index):
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return int(distance_matrix[from_node][to_node])

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    # demand callback
    def demand_callback(from_index):
        from_node = manager.IndexToNode(from_index)
        return demands[from_node]

    demand_callback_index = routing.RegisterUnaryTransitCallback(demand_callback)

    # capacity dimension
    routing.AddDimensionWithVehicleCapacity(
        demand_callback_index,
        0,  # geen overschrijding toegestaan
        vehicle_capacities,
        True,
        "Capacity"
    )

    return manager, routing
```

```python
# AFSTAND BEREKENEN
def calculate_total_distance(solution, routing, num_vehicles):
    total_distance = 0

    for vehicle_id in range(num_vehicles):
        index = routing.Start(vehicle_id)
        while not routing.IsEnd(index):
            next_index = solution.Value(routing.NextVar(index))
            total_distance += routing.GetArcCostForVehicle(index, next_index, vehicle_id)
            index = next_index

    return total_distance
```

```python
# ROUTES UITLEZEN
def extract_capacity_routes(solution, manager, routing, num_vehicles, locations, demands):
    totaal_afstand = 0
    routes_output = []

    for vehicle_id in range(num_vehicles):
        index = routing.Start(vehicle_id)
        route_nodes = []
        route_adressen = []
        route_distance = 0
        route_demand = 0

        while not routing.IsEnd(index):
            node = manager.IndexToNode(index)
            route_nodes.append(node)
            route_adressen.append(locations.loc[node, 'VisitingAddress'])
            route_demand += demands[node]

            previous_index = index
            index = solution.Value(routing.NextVar(index))
            route_distance += routing.GetArcCostForVehicle(previous_index, index, vehicle_id)

        # einddepot toevoegen
        end_node = manager.IndexToNode(index)
        route_nodes.append(end_node)
        route_adressen.append(locations.loc[end_node, 'VisitingAddress'])

        totaal_afstand += route_distance

        routes_output.append({
            "vehicle_id": vehicle_id,
            "route_nodes": route_nodes,
            "route_adressen": route_adressen,
            "aantal_adressen": route_demand,
            "afstand_meters": route_distance,
            "afstand_km": round(route_distance / 1000, 2)
        })

    return totaal_afstand, routes_output
```

```python
# FUNCTIE: ALLE COMBINATIES TESTEN VOOR MODEL 2
def compare_capacity_strategies(distance_matrix,
                                num_vehicles,
                                demands,
                                vehicle_capacities,
                                first_solution_strategies,
                                local_search_metaheuristics,
                                time_limit_seconds=10):
    results = []

    for fs_name, fs_value in first_solution_strategies.items():
        for ls_name, ls_value in local_search_metaheuristics.items():

            # vers model per combinatie
            manager, routing = build_capacity_model(
                distance_matrix=distance_matrix,
                num_vehicles=num_vehicles,
                demands=demands,
                vehicle_capacities=vehicle_capacities,
                depot=0
            )

            search_parameters = pywrapcp.DefaultRoutingSearchParameters()
            search_parameters.first_solution_strategy = fs_value
            search_parameters.local_search_metaheuristic = ls_value
            search_parameters.time_limit.seconds = time_limit_seconds

            start = time.time()
            solution = routing.SolveWithParameters(search_parameters)
            runtime = time.time() - start

            if solution:
                total_distance = calculate_total_distance(solution, routing, num_vehicles)

                # optioneel: aantal gebruikte voertuigen tellen
                used_vehicles = 0
                for vehicle_id in range(num_vehicles):
                    index = routing.Start(vehicle_id)
                    next_index = solution.Value(routing.NextVar(index))
                    if not routing.IsEnd(next_index):
                        used_vehicles += 1

                results.append({
                    "first_solution_strategy": fs_name,
                    "local_search_metaheuristic": ls_name,
                    "solution_found": True,
                    "total_distance": total_distance,
                    "used_vehicles": used_vehicles,
                    "runtime_sec": round(runtime, 2),
                })
            else:
                results.append({
                    "first_solution_strategy": fs_name,
                    "local_search_metaheuristic": ls_name,
                    "solution_found": False,
                    "total_distance": None,
                    "used_vehicles": None,
                    "runtime_sec": round(runtime, 2),
                })

    df_results = pd.DataFrame(results)

    df_results = df_results.sort_values(
        by=["solution_found", "total_distance", "runtime_sec"],
        ascending=[False, True, True]
    ).reset_index(drop=True)

    return df_results
```

```python
# BESTE PARAMETERS KIEZEN
def get_best_parameters(df_results, allow_zero_distance=False):
    df_valid = df_results[df_results["solution_found"] == True].copy()
    df_valid = df_valid[df_valid["total_distance"].notna()]

    if not allow_zero_distance:
        df_valid = df_valid[df_valid["total_distance"] > 0]

    if df_valid.empty:
        raise ValueError(
            "Geen geldige oplossing gevonden voor model 2. "
            "Controleer of er genoeg voertuigen zijn voor de capaciteitsrestrictie."
        )

    best_row = df_valid.sort_values(
        by=["total_distance", "runtime_sec"],
        ascending=[True, True]
    ).iloc[0]

    return best_row["first_solution_strategy"], best_row["local_search_metaheuristic"]

```

```python
# COMBINATIES TESTEN VOOR MODEL 2
df_results_model2 = compare_capacity_strategies(
    distance_matrix=distance_matrix,
    num_vehicles=num_vehicles,
    demands=demands,
    vehicle_capacities=vehicle_capacities,
    first_solution_strategies=first_solution_strategies,
    local_search_metaheuristics=local_search_metaheuristics,
    time_limit_seconds=TIME_LIMIT_SECONDS
)

print("Top 10 beste combinaties voor MODEL 2:")
print(df_results_model2.head(10))
```

```python
# BESTE COMBINATIE KIEZEN
best_fs2, best_ls2 = get_best_parameters(df_results_model2, allow_zero_distance=False)

print("\nBeste combinatie voor MODEL 2:")
print(f"FirstSolutionStrategy: {best_fs2}")
print(f"LocalSearchMetaheuristic: {best_ls2}")
```
