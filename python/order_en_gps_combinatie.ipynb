```python
# bekijken welke auto's elke dag naar het depot gaan
depot_naam = "Kamerlingh Onnesweg 9, Dordrecht"
```

```python
# eerste stop per dag per voertuig
idx_eerste = df.groupby(["Date", "SerialNumber"])["Start"].idxmin()
eerste_stop = df.loc[idx_eerste, ["Date", "SerialNumber", "VisitingAddress"]]
eerste_stop = eerste_stop.rename(columns={"VisitingAddress": "eerste_stop"})

# laatste stop per dag per voertuig
idx_laatste = df.groupby(["Date", "SerialNumber"])["Stop"].idxmax()
laatste_stop = df.loc[idx_laatste, ["Date", "SerialNumber", "VisitingAddress"]]
laatste_stop = laatste_stop.rename(columns={"VisitingAddress": "laatste_stop"})

# samenvoegen
start_eind_check = eerste_stop.merge(
    laatste_stop,
    on=["Date", "SerialNumber"],
    how="inner"
)

# alleen groepen die starten én eindigen bij depot
groepen_depot_start_eind = start_eind_check[
    (start_eind_check["eerste_stop"] == depot_naam) &
    (start_eind_check["laatste_stop"] == depot_naam)
][["Date", "SerialNumber"]]

# originele dataset filteren op die groepen
df01 = df.merge(
    groepen_depot_start_eind,
    on=["Date", "SerialNumber"],
    how="inner"
)

df01
```

```python
df01["adres"] = df01["VisitingAddress"].str.split(",").str[0]
```

```python
df01 = df01.rename(columns={
    "ZipCode": "postcode",
    "City": "plaats",
    "Lat": "latitude",
    "Long": "longitude"
})
```

```python
df02["adres"] = df02["adres"].str.strip().str.lower()
df02["postcode"] = df02["postcode"].str.strip().str.upper()
df02["plaats"] = df02["plaats"].str.strip().str.lower()
```

```python
df02["postcode"] = (
    df02["postcode"]
    .str.strip()
    .str.upper()
    .str.replace(" ", "", regex=False)
)
```

```python
df02["postcode"] = df02["postcode"].str.replace(
    r"(\d{4})([A-Z]{2})", r"\1 \2", regex=True
)
```
