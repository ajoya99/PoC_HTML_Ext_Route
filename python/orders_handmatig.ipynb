## GPS data

```python
df_route = pd.read_csv(
    'data/df04_latest.csv')
```

```python
df_route
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Voorbeeld: df heeft kolommen 'lat' en 'lon'
# df = pd.read_csv("data.csv")

gdf = gpd.GeoDataFrame(
    df_route,
    geometry=gpd.points_from_xy(df_route["Long"], df_route["Lat"]),
    crs="EPSG:4326"  # WGS84, standaard voor lat/lon
)

# Contextily werkt het best in Web Mercator
gdf_3857 = gdf.to_crs(epsg=3857)

ax = gdf_3857.plot(
    figsize=(8, 10),
    markersize=20,
    alpha=0.7
)

ctx.add_basemap(
    ax,
    source=ctx.providers.OpenStreetMap.Mapnik
)

ax.set_title("Punten op kaart van Nederland")
ax.set_axis_off()

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

gdf = gpd.GeoDataFrame(
    df_route,
    geometry=gpd.points_from_xy(df_route["Long"], df_route["Lat"]),
    crs="EPSG:4326"
).to_crs(epsg=3857)

ax = gdf.plot(
    figsize=(10, 10),
    markersize=35,
    alpha=0.8,
    edgecolor="black",
    linewidth=0.4
)

ctx.add_basemap(
    ax,
    source=ctx.providers.CartoDB.Positron
)

ax.set_title("Locaties op kaart", fontsize=16)
ax.set_axis_off()
plt.show()
```
