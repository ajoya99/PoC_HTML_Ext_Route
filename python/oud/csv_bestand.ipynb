```python
def maak_korte_naam(naam):
    if pd.isna(naam):
        return None
    delen = naam.lower().strip().split()
    if len(delen) >= 2:
        return delen[0] + ' ' + delen[1][0]
    elif len(delen) == 1:
        return delen[0]
    return None

df_ritten['Driver_short'] = df_ritten['DriverName'].apply(maak_korte_naam)
```

```python
df04['Monteur_short'] = (
    df04['Vaste monteur']
    .astype(str)
    .str.lower()
    .str.strip()
)
```

```python
df_ritten_1 = df_ritten[
    df_ritten['Driver_short'].isin(df04['Monteur_short'])
]
```

```python
# Veilig kopieën maken
df_orders = df05.copy()
df_ritten = df_ritten_1.copy()

# Datumkolommen naar alleen datum/tijd 00:00:00
df_orders['Responsdatum_dag'] = df_orders['Responsdatum'].dt.normalize()
df_ritten['Date_dag'] = df_ritten['Date'].dt.normalize()

# Tekstkolommen opschonen
df_orders['Plaats_clean'] = df_orders['Plaats'].str.strip().str.lower()
df_orders['Adres_clean'] = df_orders['Adres'].str.strip().str.lower()
df_orders['Monteur_short_clean'] = df_orders['Monteur_short'].str.strip().str.lower()

df_ritten['Plaats_bestemming_clean'] = df_ritten['Plaats bestemming'].str.strip().str.lower()
df_ritten['Adres_bestemming_clean'] = df_ritten['Adres bestemming'].str.strip().str.lower()
df_ritten['Driver_short_clean'] = df_ritten['Driver_short'].str.strip().str.lower()

# Merge
df_merge = df_ritten.merge(
    df_orders,
    left_on=[
        'Date_dag',
        'Plaats_bestemming_clean',
        'Adres_bestemming_clean',
        'Driver_short_clean'
    ],
    right_on=[
        'Responsdatum_dag',
        'Plaats_clean',
        'Adres_clean',
        'Monteur_short_clean'
    ],
    how='left'
)

# Check hoeveel matches
print(df_merge['Responsdatum'].notna().value_counts())
print('Aantal matches:', df_merge['Responsdatum'].notna().sum())
print('Matchpercentage:', round(df_merge['Responsdatum'].notna().sum() / len(df_merge) * 100, 2), '%')
```
