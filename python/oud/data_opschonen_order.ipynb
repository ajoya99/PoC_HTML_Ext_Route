```python
df_fleetgo = pd.read_excel("Bezoeken per voertuig 01-01-2025 - 31-01-2025.xlsx")
```

```python
df20 = df_fleetgo[df_fleetgo['GroupName'] == '19210700 Dordrecht - BBT']
df20.info()
```

```python
df21 = df20.drop(columns=['Description', 'Contact', 'Remark', 'ProjectNumber'])
df21.info()
```

```python
df21['DriverName'].unique()
```

```python
conflicten = (
    df21.assign(Naam_monteur_schoon=df21['DriverName'].replace('', pd.NA))
      .dropna(subset=['Naam_monteur_schoon'])
      .groupby(['Date', 'SerialNumber'])['Naam_monteur_schoon']
      .agg([
          ('aantal_unieke_monteurs', 'nunique'),
          ('monteurs', lambda x: list(pd.unique(x)))
      ])
      .reset_index()
)

conflicten = conflicten[conflicten['aantal_unieke_monteurs'] > 1]

print(conflicten)
```

```python
df21[df21['DriverName'] == 'Auto Mertens Auto Mertens']
```

```python
# Auto Mertens eruit halen
df22 = df21.drop(df21[df21['DriverName'] == 'Auto Mertens Auto Mertens'].index)
df22['DriverName'].unique()
```

```python
# nu kijken naar NaN waarden
df22[df22['DriverName'].isna()]
```

```python
df22.groupby(['Date', 'SerialNumber'])['DriverName'].unique()
```

```python
# lege strings eerst als missend behandelen
df22['DriverName'] = df22['DriverName'].replace('', pd.NA)

# vul alleen groepen zonder conflict
df22['DriverName'] = (
    df22.groupby(['Date', 'SerialNumber'])['DriverName']
      .transform(lambda x: x.ffill().bfill() if x.dropna().nunique() <= 1 else x)
)

df22['DriverName'].unique()
```

```python
df22[df22['DriverName'].isna()]
```

```python
df22[
    df22.groupby(['Date', 'SerialNumber'])['DriverName']
        .transform(lambda x: x.isna())
]
```

```python
# dit zijn dus rijen die bij het kenteken en de datum geen driver hebben gevonden en deze halen we er dus uit. en zijn ook de enige rijen die nog Nan zijn
df23 = df22.drop(df22[
    df22.groupby(['Date', 'SerialNumber'])['DriverName']
        .transform(lambda x: x.isna())
].index)
```

```python
df23['DriverName'].unique()
```

```python
# namen omzetten naar alleen voornaam en eerste letter van achternaam
df23['Naam monteur'] = df23['DriverName'].str.split().str[0] + ' ' + df23['DriverName'].str.split().str[-1].str[0]
```

```python
df23['Naam monteur'].unique()
```

```python
# nu van orders de namen hetzelfde krijgen
df_orders['Vaste monteur'].unique()
```

```python
df_orders['Vaste monteur'] = df_orders['Vaste monteur'].replace('', pd.NA)
```

```python
mapping = (
    df_orders.dropna(subset=['Vaste monteur'])
      .groupby('Zorginstellingnaam')['Vaste monteur']
      .agg(lambda x: x.value_counts().idxmax())
)
mapping
```

```python
df_orders['Vaste monteur'] = df_orders.apply(
    lambda row: mapping.get(row['Zorginstellingnaam'], pd.NA)
    if pd.isna(row['Vaste monteur'])
    else row['Vaste monteur'],
    axis=1
)
```
